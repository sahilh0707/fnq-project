```python
#using track 1
import yfinance as yf #in this code yfinance was giving error for downloading data on jupyter notebook, so i switched to google colab and it started working...
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from datetime import datetime,timedelta
```

```python
ticker=['NESTLEIND.NS']
enddate=datetime.today()
print(enddate)
```

```python
startdate=enddate-timedelta(days=1*90)
print(startdate)
```

```python
data=yf.download(ticker,start=startdate,end=enddate)
data
```

```python
def simplemovingaverage(info,period,column='Close'):
    return info[column].rolling(window=period).mean()
def standarddeviation(info,period,column='Close'):
    return info[column].rolling(window=period).std()
```

```python
data['SMA']=simplemovingaverage(data,20)               #SimpleMovingAverage
data['STD']=standarddeviation(data,20)

                                                                                          #X=current value(Close),U = mean value(SMA),STD=standard deviation
data[('Z_score', '')] =(data[('Close', 'NESTLEIND.NS')]-data[('SMA', '')])/data[('STD','')]       #Z_score((X-U)/STD) tells you how far a value is from the mean value
                                                                                          #as in mean reversion basic idea is to buy a stock if it is below its mean predicticting it will go back to its mean value and vice vers

data
```

```python
data['Z_score'].describe()
```

```python
#ploting z_scores
plt.figure(figsize=(15,7))
plt.title('z_scores')
plt.plot(data.index,data[('Z_score','')],label='Z_score')
```

```python
#generating signals(tells whether to buy or sell)
data[('Signal','')]=0        #initialising signal
data.loc[data['Z_score','']<-1,('Signal','')]=1         #ENTRY CONDITION-buy signal as Z_score is less,price is below avg price, hence it may revert back to its mean position acc to mean reversion method
data.loc[data['Z_score','']>1,('Signal','')]=-1         #EXIT CONDITION-sell signal as z_score is more,price is above avg price, hence it may revert back to its mean position
data.loc[(data['Z_score'] >= -1) & (data['Z_score'] <= 1), 'Signal'] = 0
```

```python
#plotting buy and sell signals on the above graph
plt.figure(figsize=(15,7))
plt.plot(data.index,data[('Z_score','')],label='Z_score')       #this will make full graph of z_score with time(2020-now)
plt.axhline(1,color='red',linestyle='--',label='OverBought, Z>1')
plt.axhline(-1,color='green',linestyle='--',label='OverSold,Z<-1')
plt.axhline(0,color='black')

#marking signals in graph
plt.scatter(data.index[data[('Signal','')] == 1],data[('Z_score','')][data[('Signal','')]==1],label='Buy Signal',marker='^',color='green')
plt.scatter(data.index[data[('Signal','')] == -1],data[('Z_score','')][data[('Signal','')]==-1],label='Sell Signal',marker='v',color='red')
plt.legend()#it adds a little box in graph showing what each line depict
plt.title('Z_Score with signals')
```

```python
#backtest- basically checking how much money would i have made if i try the same startegy before a certain period of time on my data,doing it using manual method/basic code, not using library
def backtest(info,money):             #not considering transaction taxes and commisions

    Smoney=money
    no_stocks=0
    market_money=[]           #portfolio
    cash=Smoney

    for i in range(90):       #we took data of 3 months initially
        if info[('Signal','')].iloc[i]==1 and no_stocks==0:
            no_stocks=cash/info[('Close','')].iloc[i]
            cash=0
        elif info[('Signal','')].iloc[i]==-1 and no_stocks>0:
            cash=no_stocks*info[('Close','')].iloc[i]
            no_stocks=0
        market_money.append(cash+no_stocks*info[('Close','')].iloc[i])
    info[('market_money','')]=market_money
    return info
```

```python
#running our backtest
backtest(data,1000)           #getting indexing error
```

```python

#brollinger bands(consists of 3 bands, upper,lower and middle band),MB=Simple Moving Avg,(UB,LB=MB+-(2*SD) resp)
```

```python
data['UB','']=data['SMA','']+2*data['STD','']
data['LB','']=data['SMA','']-2*data['STD','']
```

```python
data
```

```python
#generating signals for brollinger bands
# Initializing the 'Signal' column to 0 (no signal)
close=data['Close']
lb=data['LB']
ub=data['UB']

close, lb = close.align(lb, join='inner', axis=0)  # Align on index and remove rows that don't align
close, ub = close.align(ub, join='inner', axis=0)  #

data['SignalB'] = 0
data.loc[close < lb, 'SignalB'] = 1  # Buy signal
data.loc[close > ub, 'SignalB'] = -1  # Sell signal
```

```python
plt.figure(figsize=(15,7))
plt.plot(data.index,data[('lb','')],label='Lower Band',color='black')
plt.plot(data.index,data[('ub','')],label='Upper Band',color='red')
plt.plot(data.index,data[('SMA','')],label='SMA',color='steel blue')
plt.plot()
```
